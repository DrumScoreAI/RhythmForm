```python
import pandas as pd
import os
import glob
import numpy as np
import json
import matplotlib.pyplot as plt
```

```python
smt_pred_dir = os.path.abspath('../temp/smt_predictions')
os.listdir(smt_pred_dir)
```

```python
csvs = []
for csvf in glob.glob(os.path.join(smt_pred_dir, 'prediction*.csv')):
    csvs.append(pd.Series(pd.read_csv(csvf, header=None).values.flatten()))
```

```python
cols = np.arange(max([len(c) for c in csvs]))
cols
```

```python
smt_df = pd.DataFrame(csvs, columns=cols)
```

```python
smt_df
```

```python
vocabf = os.path.abspath('../training_data/merged_tokenizer_vocab.json')
with open(vocabf, 'r') as f:
    vocab = json.load(f)
inv_vocab = {v: k for k, v in enumerate(vocab)}
inv_vocab
```

```python
smt_map = smt_df.map(lambda x: inv_vocab.get(x, 9999))
smt_map
```

```python
gt = pd.Series(pd.read_csv(smt_pred_dir + '/itsabeautifulday.csv', header=None).values.flatten())[:smt_df.shape[1]]
```

```python
gt.head(20)
```

```python
gt_map = gt.map(lambda x: inv_vocab.get(x, 9999))
gt_map
```

```python
def highlight_diff(row):
    """Highlights differences between a row and the ground truth series 'gt'."""
    gt_aligned = gt_map.reindex(row.index)
    is_correct = (row == gt_aligned)
    # Return a list of styles: green for correct, red for incorrect
    return ['background-color: #90EE90' if v else 'background-color: #F08080' for v in is_correct]

# Display a styled dataframe for the first 30 tokens to compare predictions.
# Correct tokens are green, incorrect ones are red.
smt_map.iloc[:, :100].style.apply(highlight_diff, axis=1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a dataframe indicating correctness (True/False) up to the length of the ground truth.
correctness_df = smt_map.iloc[:, :len(gt_map)].eq(gt_map, axis=1)

# Create the heatmap visualization.
plt.figure(figsize=(20, 10)) # Use a large figure size to see as much detail as possible.
sns.heatmap(correctness_df, cbar=False, cmap=['#000000', '#FFFFFF']) # Black for incorrect, White for correct.
plt.title('Prediction Correctness Heatmap (White = Correct)', fontsize=16)
plt.xlabel('Token Position', fontsize=12)
plt.ylabel('Epoch', fontsize=12)
plt.show()
```

```python
# Calculate the token-level accuracy for each epoch by comparing smt_df with gt.
# This comparison relies on smt_df and gt having the same length and alignment.
accuracy = (smt_df.eq(gt, axis=1)).sum(axis=1) / len(gt)

# Create the plot
plt.figure(figsize=(14, 7))
accuracy.plot(kind='line', marker='o', linestyle='-')
plt.title('Token Accuracy vs. Ground Truth per Epoch', fontsize=16)
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.grid(axis='y', linestyle='--')
plt.ylim(0, max(1.0, accuracy.max() * 1.1)) # Set y-limit to 1.0 or slightly above max accuracy
plt.show()
```
